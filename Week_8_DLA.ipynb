```python
import matplotlib.pyplot as plt
plt.rcdefaults()
import numpy as np

plt.rcdefaults()
fig, ax = plt.subplots()

firms = ('Dunder Mifflin','Saber','Prince Family Paper','Georgia Pacific','UPM')
y_pos = np.arange(len(firms))
performance = 3 + 10 * np.random.rand(len(firms))
error = np.random.rand(len(firms))

ax.barh(y_pos, performance,xerr=error,align='center',
        color='purple',ecolor='grey')
ax.set_yticks(y_pos)
ax.set_yticklabels(firms)
ax.invert_yaxis()  
ax.set_xlabel('Total Sales (1000 Reams)')
ax.set_title('Q3 2004 Paper Sales')

plt.show()
```

```python
np.random.seed(19680801)
all_data = [np.random.normal(0, std, 100) for std in range(1, 4)]

fig, ax1 = plt.subplots()

bplot = ax1.boxplot(all_data,
                         notch=True,  
                         vert=True,   
                         patch_artist=True)
labels = ['x1', 'x2', 'x3']

ax1.set_title('Annoying Box Plot')


colors = ['darkgreen', 'darkblue', 'darkred']
for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)


for ax in ax1:
    ax.yaxis.grid(True)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
```
